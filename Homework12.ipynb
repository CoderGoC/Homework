# MEAN yoki MEDIAN boyicha toldiramiz Numberic datalani

```python
element_type = type(df1["yacEPA"].iloc[0])
print(element_type)
```

```python
print(df1["yacEPA"].mean())
print(df1["yacEPA"].median())
```

Median bilan toldirib chiqsak boladi


```python
df1["Median_yacEPA"] = df1["yacEPA"].fillna(value=df1["yacEPA"].median())
df1
```

```python
print(df1["airEPA"].mean())
print(df1["airEPA"].median())
```

```python
df1["Median_airEPA"] = df1["airEPA"].fillna(value=(df1["airEPA"].median()))
df1
```

Hozir shu ikkita qilgan qatorimizdagi ozgarishlarni koramiz

```python
df1["yacEPA"].hist()
```

```python
df1["Median_yacEPA"].hist()
```

```python
sns.displot(df1, x="yacEPA", kde=True, alpha=0.6)

sns.displot(df1, x="Median_yacEPA", kde=True, alpha=0.6)

plt.show()
```

```python
sns.displot(df1, x="airEPA", kde=True, alpha=0.6)

sns.displot(df1, x="Median_airEPA", kde=True, alpha=0.6)

plt.show()
```
