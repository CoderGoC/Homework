# Ishlamoqchi bolgan data frame ni kiritib olamiz

```python
Obesity = pd.read_csv("Obesity Levels.csv")
Obesity
```

### Jinslar ortasida vaznlarning taqsimlanishi


```python
sns.stripplot(data=Obesity, y="Gender",x="Weight",jitter=True)
```

Xulosa qilib aytsak, Erkaklar o'rtasida vaznlar zichligi nisbatan zichroq ayollarga qaraganda. Ayollarda tarqoqroq joylashgan ekan. Va erkaklarda bitta outlier ham mavjud ekan.

### Jinslar ortasida boylarning taqsimlanishi


```python
sns.stripplot(data=Obesity, y="Gender",x="Height",jitter=True)
```

Bundan korinadiki ayollarning ortacha boy sochilishi erkaklarnikiga qaraganda pastroq ekan. Ayollarning boylari 1,5 va 1,8 orasida bolsa erkaklarda biroz kattaroq bu korsatkich yani 1,6 dan 2,0 gachani tashkil etmoqda

### Semiz odamlarning jamoat transportidan foydalanish taqsimoti

```python
sns.stripplot(data=Obesity, y="MTRANS",x="Weight",jitter=True)
```

Bundan shuni xulosa qilsak boladiki, jismoniy harakat yordamida boshqariladigan transport vositalarni va piyodalar orasida nisbatan semizlik darajasi pastroq ekan


```python
sns.swarmplot(data=Obesity, y="Gender", x="Height",palette="viridis")
```

```python
sns.swarmplot(data=Obesity, y="Gender", x="Height", palette=["#FF6347", "#4682B4"])
```

### Bog'liqlik darajalarini miqdorlarda ko'rish

```python
heatmap_data = Obesity.pivot_table(index='Gender', columns='MTRANS', values='Weight', aggfunc='mean')


sns.heatmap(data=heatmap_data, annot=True, cmap='viridis')
```

```python
heatmap_data = Obesity.pivot_table(index='CALC', columns='MTRANS', values='Weight', aggfunc='mean')


sns.heatmap(data=heatmap_data, annot=True, cmap='viridis')
```
