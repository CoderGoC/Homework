```python
def change(num):
    x = list(str(int(num)))
    x[-1] = '0'
    return int(''.join(x))

Obesity_level['new_weight'] = Obesity_level['Weight'].agg(change)
Obesity_level['new_age'] = Obesity_level['Age'].agg(change)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.relplot(x="Weight",y="FCVC",data=Obesity_level,kind="scatter",col="Gender")
sns.relplot(x="new_weight", y="Age", data=Obesity_level, kind="line",col="Gender")


plt.show()
```

```python
male_vazn = Obesity_level[Obesity_level["Gender"]=="Male"]["Weight"].mean()
print("Erkaklar vazn - ",male_vazn)
female_vazn = Obesity_level[Obesity_level["Gender"]=="Female"]["Weight"].mean()
print("Ayollar vazn - ",female_vazn)
```

```python
male_boy = Obesity_level[Obesity_level["Gender"]=="Male"]["Height"].mean()
print("Erkaklar boyi - ",male_boy)
female_boy = Obesity_level[Obesity_level["Gender"]=="Female"]["Height"].mean()
print("Ayollar boyi - ",female_boy)
```

Bunda ko'rishimiz mumkinki, erkaklar va ayollarning o'rtacha bo'ylarini .

Erkaklar boyi -  1.76  sm

Ayollar boyi -  1.64   sm

```python
 
male_mean_height = Obesity_level[Obesity_level["Gender"] == "Male"]["Height"].mean()
female_mean_height = Obesity_level[Obesity_level["Gender"] == "Female"]["Height"].mean()
male_std_height = Obesity_level[Obesity_level["Gender"] == "Male"]["Height"].std()
female_std_height = Obesity_level[Obesity_level["Gender"] == "Female"]["Height"].std()

male_mean_weight = Obesity_level[Obesity_level["Gender"] == "Male"]["Weight"].mean()
female_mean_weight = Obesity_level[Obesity_level["Gender"] == "Female"]["Weight"].mean()
male_std_weight = Obesity_level[Obesity_level["Gender"] == "Male"]["Weight"].std()
female_std_weight = Obesity_level[Obesity_level["Gender"] == "Female"]["Weight"].std()

 
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

 
ax[0, 0].scatter(Obesity_level[Obesity_level["Gender"]=="Male"]["Height"], Obesity_level[Obesity_level["Gender"]=="Male"]["Weight"], c='blue', label="Male")
ax[0, 0].set_xlabel("Bo'y")
ax[0, 0].set_ylabel("Vazn")
ax[0, 0].legend()

ax[0, 1].scatter(Obesity_level[Obesity_level["Gender"]=="Female"]["Height"], Obesity_level[Obesity_level["Gender"]=="Female"]["Weight"], c='orange', label="Female")
ax[0, 1].set_xlabel("Bo'y")
ax[0, 1].set_ylabel("Vazn")
ax[0, 1].legend()


ax[1, 0].bar("Male", height=male_mean_height, yerr=male_std_height, color='blue', label="Male")
ax[1, 0].bar("Female", height=female_mean_height, yerr=female_std_height, color='orange', label="Female")
ax[1, 0].set_ylabel("Bo'y")
ax[1, 0].legend()

ax[1, 1].bar("Male", height=male_mean_weight, yerr=male_std_weight, color='blue', label="Male")
ax[1, 1].bar("Female", height=female_mean_weight, yerr=female_std_weight, color='orange', label="Female")
ax[1, 1].set_ylabel("Vazn")
ax[1, 1].legend()

plt.tight_layout()
plt.show()
```

Bunda biz Ayollar va erkaklarning boyi va vazni ortasidagi bogliqliklarni hamda bar graph orqali ularni alohida alohida boy va vaznlarini std larini ham korib chiqdik.
