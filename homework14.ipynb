Charges columnda yuqorida koringanidek ancha miqdorda outlierlar mavjud qoshimchasiga bu normal distributionga ham tushmayapti. Bu bizga shuni anglatadiki, agar biz har qancha outlarni yoqotsak ham yangi outlierlar kelib chiqaveradi setdan kelib chiqqan holda

Bizda hozir bmi va charges uchun chizgan boxplotlarimizi korsak ularda outlierlar mavjud va biz hozir ular bilan ishlaymiz  
Bunda bizda 2ta outlier bor colunms uchun plotlarni show qilish functionlarini yozvolamiz.

```python
def plot_bmi(Old, New):
    fig, ax = plt.subplots(ncols=2, nrows=1,figsize=(17,7))
    sns.boxplot(Old["bmi"],ax=ax[0])
    ax[0].title.set_text("BMI with outliers")
    sns.boxplot(New["bmi"],ax=ax[1])
    ax[1].title.set_text("BMI without outliers")
    ax[0].set_xlabel('BMI')
    ax[1].set_xlabel('BMI')
    
    plt.show()
  

def plot_charges(Old, New):
    fig, ax = plt.subplots(ncols=2, nrows=1,figsize=(17,7))
    sns.boxplot(Old["charges"],ax=ax[0])
    ax[0].title.set_text("Charges with outliers")
    sns.boxplot(New["charges"],ax=ax[1])
    ax[1].title.set_text("CHarges without outliers")
    ax[0].set_xlabel('Charges')
    ax[1].set_xlabel('Charges')
    
    plt.show()
    
  
def remove_outliers(data, function):
  numeric_columns = data.select_dtypes(include=['number']).columns
  copied_data = data.copy()
  for col in numeric_columns:
      lower_limit, upper_limit = function(copied_data, col)
      copied_data = copied_data[(copied_data[col] >= lower_limit) & (copied_data[col] <= upper_limit)]
  
  return copied_data


```

Biz bunda hozir har qanday data set uchun ishlatilishi mumkin bolgan fuctionlarni yozib oldik. Bular yordamida biz boxplotlarni chizamiz va outlierni bemalol remove qilsak boladi.

## Gaussian distribution (mean and std)

```python
data = pd.read_csv("insurance.csv")
def Gaussian_distribution(data, col):
    mean = data[col].mean()
    std = data[col].std()
    lower_limit = mean - 3 * std
    upper_limit = mean + 3 * std
    return lower_limit, upper_limit

data_no_outliers_gaussian = remove_outliers(data, Gaussian_distribution)
plot_bmi(data, data_no_outliers_gaussian)
plot_charges(data, data_no_outliers_gaussian)
```

#### Xulosa qiladigan bolsak bunda biz hozir, ikkala column uchun ham yozilgan functionlada outlierlar mutloqo yoqolmadi lekin ularning data set ga tasirini kopaytirmagan holda malum chegaragacha kamaytirildi. Bunda biz birinchi nazariya  â€¢ Gaussian distribution (mean and std) nazariyasi va formulalaridan foydalandik va keraklik natijalarga qisman erishdik. Data setni ancha normallashtirdik


### IQR proximity rule

```python
data = pd.read_csv("insurance.csv")

def IQR_proximity_rule(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    return lower_limit, upper_limit

data_no_outliers_iqr = remove_outliers(data, IQR_proximity_rule)

plot_bmi(data, data_no_outliers_iqr)
plot_charges(data, data_no_outliers_iqr)
```

#### Bunda ham huddi shunday hulosaga kelsak boladi, yani set dagi malum bir outlierlar IQR proximity rule metodi asosida chiqarib yuborildi.  Charges columnda data miqdori ancha kop va outlierlar ham yetarli darajada kopligini hisobga olgan holda max darajada outlierlar qisqartirildi va set normallashtirildi

### Quantiles (5% and 95%)

```python
data = pd.read_csv("insurance.csv")


def Quantiles(data, col):
    lower_limit = data[col].quantile(0.05)
    upper_limit = data[col].quantile(0.95)
    return lower_limit, upper_limit
  
data_no_outliers_quantiles = remove_outliers(data, Quantiles)
plot_bmi(data, data_no_outliers_quantiles)
plot_charges(data, data_no_outliers_quantiles)
```

#### Quantiles methodida ham outlierlar miqdori kamayptirilda va set normallashtirildi ancha
