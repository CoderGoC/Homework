```python
df["DEBT_MDN_FILLNA_MEAN"] = (
    df
    .groupby(by=["REGION","Ownership"])["DEBT_MDN_numeric"]
    .transform(lambda x: x.fillna(x.mean()))
)

```

```python
df["DEBT_MDN_FILLNA_MEAN"].isnull().sum()
```

```python
df["DEBT_MDN_FILLNA_MEAN"].info()
```

Korib turganizdek NaN yoki str typelar qolmadi 

```python
df["COMPL_RPY_5YR_RT_FILLNA_MEAN"] = (
    df
    .groupby(by=["REGION","Ownership"])["COMPL_RPY_5YR_RT_numeric"]
    .transform(lambda x: x.fillna(x.mean()))
)

```

```python
df["COMPL_RPY_5YR_RT_FILLNA_MEAN"].isnull().sum()
```

```python
df["COMPL_RPY_5YR_RT_FILLNA_MEAN"].info()
```

```python
df["MEDIAN_HH_INC_FILLNA_MEAN"] = (
    df
    .groupby(by=["REGION","Ownership"])["MEDIAN_HH_INC_numeric"]
    .transform(lambda x: x.fillna(x.mean()))
)

```

```python
df["MEDIAN_HH_INC_FILLNA_MEAN"].isnull().sum()
```
