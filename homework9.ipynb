```python
sns.set_style("whitegrid") 


sns.catplot(x="MTRANS", data=Obesity, hue="Gender",kind="count") 

plt.xticks(rotation=30)

plt.show()
```

Bu jadval ayollar va erkaklarning public transportlarni qaysi birini qancha foydalanishini ifodalaab keladi

```python
sns.set_style("whitegrid")
sns.set_context("notebook")


sns.catplot(x="MTRANS", y="Age", data=Obesity, hue="Gender",kind="point") 

plt.xticks(rotation=30) 

plt.grid(True)

plt.show()
```

BUnda biz orqa dizaynga set_context methodi orqali kataklarni qoshib oldik

CATPLOT chizib koramiz

```python


plot = sns.catplot(x="MTRANS", y="Age", data=Obesity, hue="Gender", kind="box")

plt.xticks(rotation=45)

plot.figure.suptitle("New Title")

plt.tight_layout()

plt.show()
```

Xulosa qilib aytsak, har bir turda deyarlik outlier mavjud. Eng kop foydalanuvchilar ayollar va erkaklar uchun ham avtomobil togri kelmoqda

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.set_palette(["blue", "red"])

plot = sns.catplot(x="MTRANS", y="Age", data=Obesity, hue="Gender", kind="point", col = "family_history_with_overweight")

for ax in plot.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
    


plot.figure.suptitle("Survey about ovearweight")

plt.tight_layout()

plt.show()
```

bunda semizlar historysini korsak ham korinib turibdiki, semizlar kam bolgan oilalar doimiy ravishda public transportlaridan kam foydalanib kelishgan.

Endi xuddi shu natijani chekuvchi va chekmaydiganlar uchun koramiz

```python
sns.set_style("whitegrid")
sns.set_context("notebook")
sns.set_palette(["blue", "red"])

plot = sns.catplot(x="MTRANS", y="Age", data=Obesity, hue="Gender", kind="point", col = "SMOKE")

for ax in plot.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
    


plot.figure.suptitle("Survey about ovearweight")

plt.tight_layout()

plt.show()
```
